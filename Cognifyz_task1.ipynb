# Preprocessing

Handling missing values

```python
df['Cuisines'].fillna('Unknown', inplace=True)

df['Cuisines'].head(89) #NaN replaced by Unknown
```

Feature Extraction

```python
x=df.drop('Aggregate rating',axis=1)
y= df['Aggregate rating']

categorical_features=x.select_dtypes(include=['object']).columns
categorical_features
```

```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
numerical_features = numerical_features.drop('Aggregate rating')
```

Preprocessing



```python
preprocessor = ColumnTransformer(
    transformers=[('cat',OneHotEncoder(handle_unknown='ignore'),categorical_features),
                  ('num', 'passthrough', numerical_features)  # 'passthrough' ensures numerical features are not dropped
                  ]
)
```

Splitting Dataset

```python
# Separate features and target variable
X = df.drop('Aggregate rating', axis=1)
y = df['Aggregate rating']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=11)
```

```python
#columns on which model is trained
X
```

## Model Evaluation and Selection

```python
models = {
    'Linear Regression' : LinearRegression(),
    'Decision Tree' : DecisionTreeRegressor(random_state=11),
    'Random Forest' : RandomForestRegressor(random_state=11),
    'Gradient Boost' : GradientBoostingRegressor(random_state=11),
}

results = {}

for name,model in models.items():
  pipeline = Pipeline(
      steps=[
          ('preprocessor',preprocessor),
          ('regressor', model)
      ]
  )

  pipeline.fit(x_train,y_train)
  y_pred=pipeline.predict(x_test)
  mse = mean_squared_error(y_test,y_pred)
  r2 = r2_score(y_test,y_pred)
  results[name] = {'MSE' : mse , 'R2' : r2}

for name,metrics in results.items():
  print(f"{name} - MSE : {metrics['MSE']}, R2: {metrics['R2']}")
```

```python
best_model = GradientBoostingRegressor(random_state=11)
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', best_model)
])

pipeline.fit(x_train, y_train)
```

# Most Influential OneHot Encoded Feature



```python
# Get feature importances from the Gradient Boosting model
importances = pipeline.named_steps['regressor'].feature_importances_

# Get the feature names after preprocessing
cat_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
num_feature_names = list(numerical_features)
feature_names = list(cat_feature_names) + num_feature_names

# Create a DataFrame to display feature importances
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

# Sort the DataFrame by importance for better readability
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print(importance_df)

# Get the most important feature
most_important_feature = importance_df.iloc[0]
print(f"The most important feature is: {most_important_feature['Feature']} with importance: {most_important_feature['Importance']}")
```

# Most Influential Original Feature



```python
# Create a mapping from original feature names to preprocessed feature names
original_to_processed_mapping = {}
for original_feature in df.drop('Aggregate rating', axis=1).columns:
    for processed_feature in feature_names:
        if processed_feature.startswith(original_feature):
            original_to_processed_mapping[original_feature] = processed_feature
            break


# Map feature importances back to original feature names
importance_dict = {}
for feature, importance in zip(feature_names, importances):
    for original_feature, processed_feature in original_to_processed_mapping.items():
        if feature == processed_feature:
            importance_dict[original_feature] = importance
            break

# Create a DataFrame to display feature importances
importance_df = pd.DataFrame({
    'Feature': list(importance_dict.keys()),
    'Importance': list(importance_dict.values())
})

# Sort the DataFrame by importance for better readability
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print(importance_df)

# Get the most important feature
most_important_feature = importance_df.iloc[0]
print(f"The most important feature is: {most_important_feature['Feature']} with importance: {most_important_feature['Importance']}")
```

```python
# Assuming importance_df is your DataFrame containing feature importances
importance_df_normalized = importance_df.copy()  # Make a copy to avoid modifying the original


importance_df_normalized['Importance_normalized'] = (importance_df_normalized['Importance'] - importance_df_normalized['Importance'].min()) / \
                                                     (importance_df_normalized['Importance'].max() - importance_df_normalized['Importance'].min())

print(importance_df_normalized)
```
